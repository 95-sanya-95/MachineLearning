## Giving the necessary details for HOG method

```python
img = cv2.imread('/content/footballer-cut-out-white-19409785.jpg.webp')
```

```python
image_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) ## As the HOG method expects grayscale image
```

```python
# Calculate HOG features
hog_features, hog_image = hog(image_gray, orientations=8, pixels_per_cell=(16, 16),
                              cells_per_block=(1, 1), visualize=True, block_norm='L2-Hys')
```

```python
# Rescale histogram for better visualization
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 5)) ## the choice of the specific range (0, 5)
#  s arbitrary and can be adjusted based on the desired contrast enhancement and visualization requirements.
```

```python
plt.imshow(hog_image_rescaled)
plt.show()
```
