```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('USA_cars_datasets-2.csv')
df
```

## **Data Cleaning**

As concluded from the last colab https://colab.research.google.com/drive/1rMoplAtMiNwbjCPiuGT5IsNo578lXNVN#scrollTo=TrdPbt9cfP8Y

I will drop all the columns except those on which price was dependent

```python
df = df.drop(['Unnamed: 0','mileage','vin','lot','condition'], axis='columns')
df.head()
```

# **Feature Engineering**

Since the machine model is comfortable in having the independent variables as numeric values, we need to change the variables such as brand, model, title_status, color, state and country in numeric values.

This will be done using ***One-Hot encoding*** and ***Dummy Variables***

### **DUMMY VARIABLES**

```python
dummy1  = pd.get_dummies(df.brand)
dummy1
```

```python
dummy2 = pd.get_dummies(df.model)
dummy2
```

```python
dummy3 = pd.get_dummies(df.title_status)
dummy3
```

```python
dummy4= pd.get_dummies(df.color)
dummy4
```

```python
dummy5 = pd.get_dummies(df.state)
dummy5
```

```python
dummy6 = pd.get_dummies(df.country)
dummy6
```

```python
merged = pd.concat([df,dummy1, dummy2,dummy3,dummy4,dummy5,dummy6], axis='columns')
merged
```

```python
final = merged.drop(['brand','model','title_status','color','state','country'], axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
X=final.drop('price', axis='columns')
X
```

```python
y=final['price']
y
```

```python
model.fit(X,y)
```

```python
model.predict(X)
```

### Now we will check how accurate our model is which can be done using ***score*** function.

```python
model.score(X,y)
```

This shows that the model is 72.80% accurate
