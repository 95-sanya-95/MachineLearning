```python
import cv2
```

## **Implementing Haar_cascade available on opencv github page**

```python
face_cascade = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
```

here we have used CascadeClassifier which is a ***face-detection algorithm***

```python
import matplotlib.pyplot as plt
import matplotlib.image as img
image = img.imread('/content/group-of-young-adults-diverse-friends-732x549-thumbnail.jpg.webp')
plt.imshow(image)
plt.show()
```

```python
image2 = img.imread('/content/group-of-young-adults-diverse-friends-732x549-thumbnail.jpg.webp')
```

```python
image3 = img.imread('/content/group-of-young-adults-diverse-friends-732x549-thumbnail.jpg.webp')
```

# **Understanding CascadeClassifier**

Here we will be using ***detectMultiscale*** method of CascadeClassifier which requires the input image to be ***gray***

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
gray2= cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
```

```python
gray3 = cv2.cvtColor(image3, cv2.COLOR_BGR2GRAY)
```

face_cascade.detectMultiScale(gray , X , Y) here, X is the scale factor in which the image will get reduce after each time the cascade will run and scan the entire image, so much smaller the scale factor chances of geting smaller faces will get reduced.

```python
faces = face_cascade.detectMultiScale(gray , 1.05, 4)
```

```python
faces2 = face_cascade.detectMultiScale(gray2 , 1.2, 4)
```

```python
faces3 = face_cascade.detectMultiScale(gray3 , 1.5, 4)
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(image, (x,y) , (x+w,y+h), (255,0,0),3)
```

```python
for (x,y,w,h) in faces2:
    cv2.rectangle(image2, (x,y) , (x+w,y+h), (255,0,0),3)
```

```python
for (x,y,w,h) in faces3:
    cv2.rectangle(image3, (x,y) , (x+w,y+h), (255,0,0),3)
```
